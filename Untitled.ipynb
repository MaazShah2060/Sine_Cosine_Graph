```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
a = float(input("Enter starting point(digit): "))
b = float(input("Times with pi?(Press 1 if needed else press any key to continue): "))
if(b!=1):
    b=0
c = float(input("Enter ending point(digit): "))
d = float(input("Times with pi?(Press 1 if needed else press any key to continue): "))
if(d!=1):
    d=0
x = np.arange(a*(np.pi**b),c*(np.pi**d),0.1)
i = input("What curve do you need?(*Press S for Sine and C for Cosine): ")
if(i == 's' or i =='S'):
    y = np.sin(x)
elif(i == 'c' or i == 'C'):
    y = np.cos(x)
else:
    print("INVALID ARGUMENT!!!")
plt.plot(x,y)
plt.show()
```
